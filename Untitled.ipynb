```python
def make_test_train(df):
  x_test = df[df['Month'] == df.iloc[-1]['Month']]
  x_train = df[df['Month'] != df.iloc[-1]['Month']]

  y_test = x_test['CO']
  x_test.drop(['CO'],axis=1, inplace=True)
  y_train = x_train['CO']
  x_train.drop(['CO'],axis=1, inplace=True)

  return x_train, x_test, y_train, y_test
```

```python
def MX(num_bankomat, powr=1):
    sum = 0
    k = 0
    for i in range(len(df[df['ID точки'] == str(num_bankomat)])):
        sum += df['CO'][i]**powr
        k += 1
    return sum / k
```

```python
def DX(num_bankomat):
    return MX(num_bankomat, 2) - MX(num_bankomat)**2
    
```
